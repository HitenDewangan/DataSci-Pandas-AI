##### Columns having wrong datatype
price, extra_people, security, deposit, cleaning_fee

```python
copy_data.drop(columns = ['experiences_offered','has_availability'], axis=1, inplace = True)  #removing cols having no need
```

```python
copy_data.has_availability.unique()
```

```python
copy_data.price = copy_data.price.apply(lambda x : x.replace("$", ""))   #remove dollar sign
copy_data['price'] = copy_data['price'].str.replace("$", "")
```

```python
#similarly do for other cols
cols_to_treat = ['price','extra_people', 'security_deposit', 'cleaning_fee']

for i in cols_to_treat:
    # Remove '$' from the specified columns
    copy_data[i] = copy_data[i].astype(str).str.replace("$", "")
    
    # Convert the columns to int64 or float64
    copy_data[i] = pd.to_numeric(copy_data[i], errors='coerce')  # Convert to numeric, handling potential errors
    copy_data[i] = copy_data[i].astype('float64')  # Convert to Int64 to handle missing values

# Now the specified columns in 'copy_data' are of type Int64 / float64
    
```

```python
copy_data[cols_to_treat].info()
```

```python
Num = []
Cat = []
```

```python
for col in copy_data.columns:
    if df[col].dtype == 'object':
        Cat.append(col)
    else:
        Num.append(col)
    
```
